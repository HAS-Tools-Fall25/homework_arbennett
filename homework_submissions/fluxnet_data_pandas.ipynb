```python
df = pd.read_csv('TimeSeries_Q_ManID_EventOnly.csv', index_col=0)
df['Date + Minute'] = pd.to_datetime(df['Date + Minute'])
site_list = df.index.unique()
```

```python
for site in site_list:
    sub = df.loc[site]
    sub.index = sub['Date + Minute']
    sub['Volume (m3)'].plot(linestyle='', marker='.', label=site)

plt.legend()
```
